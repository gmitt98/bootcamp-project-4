```python
# dependencies
import pandas as pd
import os
my_dir = os.path.dirname(os.path.abspath('__file__'))
os.chdir(my_dir)
```

```python
pdx_hpi_by_zip = 'data/cleaned data/pdx_hpi_by_zip.csv'
pdx_hpi_by_zip_df = pd.read_csv(pdx_hpi_by_zip)

PDX_Crime_zip_year = 'data/cleaned data/PDX_Crime_zip_year.csv'
PDX_Crime_zip_year_df = pd.read_csv(PDX_Crime_zip_year)
```

```python
merged_df = pdx_hpi_by_zip_df.merge(PDX_Crime_zip_year_df, 
                                left_on=['Five-Digit ZIP Code', 'Year'],
                                right_on=['zip', 'year'],
                                how='left')
```

```python
merged_df
```

```python
columns_to_drop = ['zip_code', 'zip', 'year', 'Neighborhood', 'HPI from 2012', 'normalized_sale_price_2012']
final_df = merged_df.drop(columns=columns_to_drop)
```

```python
local_data_agg_path = 'data/cleaned data/local_data_agg.csv'
final_df.to_csv(local_data_agg_path, index = False)
```
