```python
# Dependencies and move to right folder
import pandas as pd
import os
import csv
import requests
from bs4 import BeautifulSoup
import time
my_dir = os.path.dirname(os.path.abspath('__file__'))
os.chdir(my_dir)
```

```python
headers = {
    'authority': 'www.zillow.com',
    'sec-ch-ua': '"Google Chrome";v="89", "Chromium";v="89", ";Not A Brand";v="99"',
    'sec-ch-ua-mobile': '?0',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36',
    'accept': '*/*',
    'sec-fetch-site': 'same-origin',
    'sec-fetch-mode': 'cors',
    'sec-fetch-dest': 'empty',
    'referer': 'https://www.zillow.com/los-angeles-ca/2_p/?searchQueryState=%7B%22pagination%22%3A%7B%22currentPage%22%3A2%7D%2C%22usersSearchTerm%22%3A%22Los%20Angeles%22%2C%22mapBounds%22%3A%7B%22west%22%3A-118.88620504589845%2C%22east%22%3A-117.93726095410157%2C%22south%22%3A33.54554919445917%2C%22north%22%3A34.49481997322805%7D%2C%22regionSelection%22%3A%5B%7B%22regionId%22%3A12447%2C%22regionType%22%3A6%7D%5D%2C%22isMapVisible%22%3Atrue%2C%22filterState%22%3A%7B%22sort%22%3A%7B%22value%22%3A%22globalrelevanceex%22%7D%2C%22ah%22%3A%7B%22value%22%3Atrue%7D%7D%2C%22isListVisible%22%3Atrue%7D',
    'accept-language': 'en-US,en;q=0.9',
}

base_url = 'https://www.zillow.com/'
cities = ['portland-or',
          'beaverton-or',
          'hillsboro-or',
          'tigard-or',
          'milwaukie-or',
          'lake-oswego-or',
          'troutdale-or',
          'eugene-or',
          'springfield-or',
          'corvallis-or',
          'salem-or',
          'vancouver-wa',
          'olympia-wa',
          'tacoma-wa',
          'seattle-wa',
          'burien-wa',
          'bothell-wa']

num_pages = 20
```

```python
for city in cities:
    url = base_url + city
    print(url)
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print (f'success: {url}')
    else:
        print(f'failure: {url}')
    time.sleep(0.5)
```

```python
df = pd.DataFrame(columns='link')
```

```python
for city in cities:
    print(f'now processing {city}')
    for i in range (1, num_pages+1):
        print(f'page... :{i}')
        time.sleep(1)
        response = requests.get(f'https://www.zillow.com/{city}/{i}_p', headers = headers)
        html = response.text
        soup = BeautifulSoup(html, "html.parser")
        all_links = soup.find_all('a', href=True)
        links = [link['href'] for link in all_links if link['href'].startswith('https://www.zillow.com/homedetails')]
        print('found ', len(links), ' links')
        for link in links:
            df.loc[len(df)] = link
```
