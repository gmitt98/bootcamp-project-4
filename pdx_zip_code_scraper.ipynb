```python
# dependencies
import requests
from bs4 import BeautifulSoup
import os
import csv
my_dir = os.path.dirname(os.path.abspath('__file__'))
os.chdir(my_dir)
```

```python

# scrape the site
url = 'https://portlandneighborhood.com/portlandzipcodes'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', class_='wp-block-table')

# storing the data
table_data = []

if table:
    rows = table.find_all('tr')
    for row in rows:
        columns = row.find_all('td')
        row_data = [column.get_text(strip=True) for column in columns]
        # delete text "View the" and "Neighborhood profile"
        row_data[0] = row_data[0].replace('View the', '').replace('Neighborhood profile', '')
        table_data.append(row_data)
else:
    print("Table not found.")

```

```python
# check the data
for row in table_data:
    print(row)
```

```python
csv_filename = 'data/uncleaned data/local data/pdx_zip_codes_scraped.csv'
# save to csv
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['neighborhood', 'zip1', 'zip2'])
    for row in table_data:
        csv_writer.writerow(row)
```
