# national data cleaning

```python
# dependencies and make sure i'm in the same dir as this notebook
import pandas as pd
import os
my_dir = os.path.dirname(os.path.abspath('__file__'))
os.chdir(my_dir)
```

```python
# clean and transform house price index
hpi_csv = 'data/uncleaned data/national data/zillow_hpi_case_shiller_us.csv'
hpi_df = pd.read_csv(hpi_csv)

hpi_df['observation_date'] = pd.to_datetime(hpi_df['observation_date'])
filtered_hpi_df = hpi_df[hpi_df['observation_date'].dt.strftime('%m-%d') == '12-01']
filtered_hpi_df['observation_date']=filtered_hpi_df['observation_date'].dt.year
filtered_hpi_df.rename(columns = {'observation_date':'year'}, inplace = True)
filtered_hpi_df['Z_YOY_change'] = filtered_hpi_df['USAUCSFRCONDOSMSAMID'].pct_change(periods=1) * 100
filtered_hpi_df['CS_YOY_change'] = filtered_hpi_df['CSUSHPINSA'].pct_change(periods=1) * 100
filtered_hpi_df.rename(columns = {'USAUCSFRCONDOSMSAMID':'zillow_index'}, inplace = True)
filtered_hpi_df.rename(columns = {'CSUSHPINSA':'case_shiller_index'}, inplace = True)
print(filtered_hpi_df)
```

```python
# clean and transform all transactions hpi
alltran_csv = 'data/uncleaned data/national data/all_transactions_hpi_us.csv'
alltran_df = pd.read_csv(alltran_csv)

alltran_df['observation_date'] = pd.to_datetime(alltran_df['observation_date'])
filtered_alltran_df = alltran_df[alltran_df['observation_date'].dt.strftime('%m-%d') == '10-01']
filtered_alltran_df['observation_date']=filtered_alltran_df['observation_date'].dt.year
filtered_alltran_df.rename(columns = {'observation_date':'year'}, inplace = True)
filtered_alltran_df['YOY_change'] = filtered_alltran_df['USSTHPI'].pct_change(periods=1) * 100
print(filtered_alltran_df)
```

```python
# read unemployment and only keep the average
unemp_csv = 'data/uncleaned data/national data/unemployment_in_us.csv'
unemp_df = pd.read_csv(unemp_csv)
unemp_df = unemp_df[['year', 'Year_Average']]
unemp_df.rename(columns = {'Year_Average':'Unemployment_Year_Avg'}, inplace = True)
print(unemp_df)
```

```python
# read recession
recess_csv = 'data/uncleaned data/national data/recessionflag.csv'
recess_df = pd.read_csv(recess_csv)
recess_df['DATE'] = pd.to_datetime(recess_df['DATE'])
filtered_recess_df = recess_df[recess_df['DATE'].dt.strftime('%m-%d') == '10-01']
filtered_recess_df['DATE']=filtered_recess_df['DATE'].dt.year
filtered_recess_df.rename(columns = {'DATE':'year'}, inplace = True)
filtered_recess_df.rename(columns = {'JHDUSRGDPBR':'recession'}, inplace = True)
print(filtered_recess_df)
```

```python
# read federal funds rate
fed_csv = 'data/uncleaned data/national data/yearly_fed_rates_mean.csv'
fed_df = pd.read_csv(fed_csv)
fed_df.rename(columns = {'YEAR':'year'}, inplace = True)
print(fed_df)
```

```python
# merge all the data and save as csv
merged_df = pd.merge(filtered_hpi_df, filtered_alltran_df, on = 'year', how = 'outer')
merged_df = pd.merge(merged_df, unemp_df, on = 'year', how = 'outer')
merged_df = pd.merge(merged_df, filtered_recess_df, on = 'year', how = 'outer')
merged_df = pd.merge(merged_df, fed_df, on = 'year', how = 'outer')
print(merged_df)
```

```python
merged_df = merged_df.sort_values(by='year', ascending = True)
merged_df = merged_df[merged_df['year']<=2023]
```
